### Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading

```python
df = pd.read_csv('weatherHistory.csv')

df.tail(3)
```

### Year Temperature Mean

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)

df['year'] = df['Formatted Date'].dt.year
df['month'] = df['Formatted Date'].dt.month

weather_mean = df.groupby(['year','month'])['Temperature (C)'].mean()
#whole data

year_mean = df.groupby('year')['Temperature (C)'].mean()
year_mean
```

### Year Summary

```python
sum_mean = df.groupby(['Summary', 'year'])['Temperature (C)'].mean()
sum_mean
```

### Ploting Temperature

```python
year_mean.plot(kind = 'bar')
plt.title('Temperature Mean')
plt.xlabel('Years')
plt.ylabel('Temperature')
```

### Ploting Summary

```python
sum_mean.plot(kind = 'line',figsize = (25,10))
plt.title('Summary mean')
plt.ylabel('Temperature')
```
